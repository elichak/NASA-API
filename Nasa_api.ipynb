```python
import requests
from PIL import Image
from io import BytesIO
import json
```

```python
key = "2Xjtli5Y7b3Y2Tn1Htk7dIKTXFdFAEDj3pEAIHZz" # ключ для доступа к NASA-API
date1 = '2022-4-5'
date2 = '2021-4-5'
date3 = '2020-4-5'
```

```python
url = "https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos?earth_date={}&api_key={}".format(date1, key)
response = requests.get(url)
file = json.loads(response.content.decode("utf-8"))
image_url = file['photos'][7]['img_src']
response = requests.get(image_url)
print(f"Фотография сделана марсоходом {file['photos'][0]['rover']['name']} в мой день рождения: 05.04.{date1[:4]}")
img = Image.open(BytesIO(response.content))
img = img.resize((700, 500))
img
```
